

*   ¬øCu√°les son las pel√≠culas de comedia con mayor score en cada a√±o?






```python
Movie_com = df[['year','score','name','genre']].copy() #Seleccionamos las columnas que necesitamos 
Movie_com_1 = Movie_com[Movie_com.genre == 'Comedy'] #Seleccionamos las peliculas de Comedia 
Movie_a√±o = Movie_com_1.year.unique() #Obtenemos los a√±os 
for n in Movie_a√±o:
  Movie_comedy_1 = Movie_com_1[Movie_com_1.year == n] #DataFrame que contiene las peliculas por a√±o
  Movie_comedy_2 = Movie_comedy_1[Movie_comedy_1.score == max(Movie_comedy_1.score)] #Contiene un DataFrame con la pelicula con el mayor puntaje obtenido en ese a√±o
  print (Movie_comedy_2)
```

```python
com = df[df.genre == 'Comedy'] #Obtenemos las peliculas de comedia
a√±o = com.year.unique() #Obtenemos los a√±os
for n in a√±o:
  comedy_1 = com[com.year == n] #DataFrame que contiene las peliculas por a√±o
  comedy_2 = comedy_1[comedy_1.score == max(comedy_1.score)] #Contiene un DataFrame con la pelicula con el mayor puntaje obtenido en ese a√±o
  print (n, comedy_2.score, comedy_2.name)

```
