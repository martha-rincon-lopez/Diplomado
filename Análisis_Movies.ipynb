

*   ¿Cuáles son las películas de comedia con mayor score en cada año?






```python
Movie_com = df[['year','score','name','genre']].copy() #Seleccionamos las columnas que necesitamos 
Movie_com_1 = Movie_com[Movie_com.genre == 'Comedy'] #Seleccionamos las peliculas de Comedia 
Movie_año = Movie_com_1.year.unique() #Obtenemos los años 
for n in Movie_año:
  Movie_comedy_1 = Movie_com_1[Movie_com_1.year == n] #DataFrame que contiene las peliculas por año
  Movie_comedy_2 = Movie_comedy_1[Movie_comedy_1.score == max(Movie_comedy_1.score)] #Contiene un DataFrame con la pelicula con el mayor puntaje obtenido en ese año
  print (Movie_comedy_2)
```

```python
com = df[df.genre == 'Comedy'] #Obtenemos las peliculas de comedia
año = com.year.unique() #Obtenemos los años
for n in año:
  comedy_1 = com[com.year == n] #DataFrame que contiene las peliculas por año
  comedy_2 = comedy_1[comedy_1.score == max(comedy_1.score)] #Contiene un DataFrame con la pelicula con el mayor puntaje obtenido en ese año
  print (n, comedy_2.score, comedy_2.name)

```



*   ¿Cuál fue el género con mayor score en 2000?



*   ¿Qué peliculas duran más de 2 horas y tienen puntuación mayor a 8.5?



```python
Movie_duracion_1 = df[['name','score','runtime']].copy()
duracion = Movie_duracion_1[(Movie_duracion_1.runtime > 120) & (Movie_duracion_1.score > 8.5)]
duracion
```



*   ¿Cuál es la pelicula con el score más bajo y más alto y qué compañía pertenecen?






```python
Movie_1 = df[['name','score','company']].copy()
Movie_min = Movie_1[Movie_1.score == min(Movie_1.score)]
Movie_min
```
